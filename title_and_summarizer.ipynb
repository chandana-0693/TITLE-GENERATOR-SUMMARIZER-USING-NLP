```python
from transformers import pipeline

# Load models
summarizer = pipeline(task="summarization", model="facebook/bart-large-cnn")
title_generator = pipeline(task="text2text-generation", model="t5-small")

print("Models Loaded Successfully")
```

```python
from transformers import pipeline

# Use flan-t5 (better than t5-small)
generator = pipeline(
    task="text2text-generation",
    model="google/flan-t5-base"
)

print("Model Loaded Successfully")
```

```python
def generate_title_summary(text):

    # Summary
    summary_prompt = "Summarize the following article clearly:\n" + text
    summary_output = generator(summary_prompt, max_length=120)
    summary_text = summary_output[0]['generated_text']

    # Title
    title_prompt = "Generate a short 5-word news headline for:\n" + text
    title_output = generator(title_prompt, max_length=12)
    title_text = title_output[0]['generated_text']

    return title_text.strip(), summary_text.strip()
```

```python
text_input = """
Natural language processing is technology that allows computers to
interpret, manipulate, and comprehend human language.
Organizations today have large volumes of voice and text data from
various communication channels like emails, text messages,
social media newsfeeds, video, audio, and more.
"""

title, summary = generate_title_summary(text_input)

print("Generated Title:")
print(title)

print("\nGenerated Summary:")
print(summary)
```
